```python
import sqlparse
```

```python
test_sql_text = """
-- My SQL code
create temp table def as 
select x,y,z from table_01;

select * from def;

with subq as (select 123 as col) select * from subq;
"""
```

```python
stmts = sqlparse.parse(test_sql_text)
print(stmts)
print(stmts[1].get_type())
```

```python
for x in stmts:
    print(x.get_type())
```
